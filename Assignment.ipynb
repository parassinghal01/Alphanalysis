```python
import pandas as pd
import yfinance as yf

# Sample input data 
data = {
    'Ticker': ['ADANIPORTS', 'APOLLOHOSP', 'ASIANPAINT'],
    'Weightage': [0.0082, 0.0061, 0.0195],
    '2023-04-27': [124.289504, 13.808715, 67.242540],
    '2023-04-28': [120.358141, 13.514411, 67.186932]
}

# Convert to DataFrame
stocks_data = pd.DataFrame(data)

# Input for the total investment amount
total_investment = 1000  # Adjust this as needed

# Calculate the number of shares for each stock on each day
results = []


start_date = input("Enter the start date (YYYY-MM-DD): ").strip()
end_date = input("Enter the end date (YYYY-MM-DD): ").strip()
total_investment = float(input("Enter the total investment amount: ").strip())

# Loop through each stock's data
for index, row in stocks_data.iterrows():
    ticker = row['Ticker']
    weightage = row['Weightage']
    
    # Calculate the investment for this stock
    investment_amount = total_investment * weightage
    
    # Loop through the dates
    for date in [start_date, end_date]:
        closing_price = row[date]
        
        # Calculate the number of shares that can be bought
        number_of_shares = investment_amount / closing_price
        
        # Append the result to the results list
        results.append({
            'Ticker': ticker,
            'Date': date,
            'Weightage': weightage,
            'Closing Price': closing_price,
        })

# Convert the results to a DataFrame
results_df = pd.DataFrame(results)

# Step 6: Pivot the DataFrame to match the output format
pivoted_df = results_df.pivot_table(index=['Ticker', 'Weightage'], columns='Date', values='Closing Price').reset_index()

# Save the results to a CSV file
pivoted_df.to_csv('Investment_Results.csv', index=False)

# Output the result for verification
pivoted_df.head()
```
